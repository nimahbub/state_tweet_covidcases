```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('state_case_tweet.csv')
```

```python
df
```

```python
# state vs number of tweets without logscale
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="state", y="num_of_tweet")
plt.title('State Vs Number of tweets [without log scale]')
plt.plot()
```

```python
# state vs number of tweets with logscale
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="state", y="num_of_tweet")
plt.yscale('log', nonposy='clip')
plt.title('State vs Number of tweets [with log scale]')

plt.plot()
```

```python
# state vs number of covid cases without logscale
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="state", y="num_of_covid_case")
plt.title('State Vs Number of covid cases [without log scale]')
plt.plot()
```

```python
# state vs number of covid cases logscale
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="state", y="num_of_covid_case")
plt.yscale('log', nonposy='clip')
plt.title('State vs Number of covid cases [with log scale]')
plt.plot()
```
